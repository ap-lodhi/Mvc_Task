# [dbo].[GeSearchDateCollegeDetails]

```sql
/****** Object:  StoredProcedure [dbo].[GeSearchDateCollegeDetails]    Script Date: 13-12-2023 11:31:06 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[GeSearchDateCollegeDetails]
    @fromDate DATE = NULL,
    @toDate DATE = NULL
AS
BEGIN
    SELECT
        cm.clg_id,
        cm.collegeName,
        cm.fees,
        cm.AddmissionDate,
        cm.Is_Payment,
        cm.Is_active,
        rm.Role_designation,
        dm.Branch_name,
        sm.Student_fisrt_Name
    FROM
        College_Master cm
    INNER JOIN
        Role_master rm ON cm.RoleDesignation = rm.Role_id
    INNER JOIN
        department_master dm ON cm.branchName = dm.Branch_id
    INNER JOIN
        student_master sm ON cm.StudenName = sm.Student_id
    WHERE
        (@fromDate IS NULL OR cm.AddmissionDate >= @fromDate)
        AND (@toDate IS NULL OR cm.AddmissionDate <= @toDate);
END;
GO
```

# [dbo].[GetCollegeFilteredData]

```sql
/****** Object:  StoredProcedure [dbo].[GetCollegeFilteredData]    Script Date: 13-12-2023 11:31:06 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[GetCollegeFilteredData]
    @fromDate DATE = NULL,
    @toDate DATE = NULL
AS
BEGIN
    SELECT
        cm.clg_id,
        cm.collegeName,
        cm.fees,
        cm.AddmissionDate,
        cm.Is_Payment,
        cm.Is_active,
        rm.Role_designation,
        dm.Branch_name,
        sm.Student_fisrt_Name
    FROM
        College_Master cm
    INNER JOIN
        Role_master rm ON cm.RoleDesignation = rm.Role_id
    INNER JOIN
        department_master dm ON cm.branchName = dm.Branch_id
    INNER JOIN
        student_master sm ON cm.StudenName = sm.Student_id
    WHERE
        cm.AddmissionDate >= CASE WHEN @fromDate IS NULL THEN cm.AddmissionDate ELSE @fromDate END
        AND cm.AddmissionDate <= CASE WHEN @toDate IS NULL THEN cm.AddmissionDate ELSE @toDate END;
END;
GO
```

```sql
/****** Object:  StoredProcedure [dbo].[ValidateLogin]    Script Date: 13-12-2023 11:31:06 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[ValidateLogin]
    @emailId NVARCHAR(255),
    @decryptedPassword NVARCHAR(50)
AS
BEGIN
    IF EXISTS (
        SELECT 1
        FROM tbl_SignUp
        WHERE emailId = @emailId
        AND decryptedPassword = @decryptedPassword
    )
    BEGIN
        SELECT
            su.id,
            su.name,
            su.emailId
        FROM tbl_SignUp su
        WHERE su.emailId = @emailId
            AND su.decryptedPassword = @decryptedPassword;
    END
END;
GO
```

# [dbo].[ValidateLoginUser]

```sql
/****** Object:  StoredProcedure [dbo].[ValidateLoginUser]    Script Date: 13-12-2023 11:31:06 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[ValidateLoginUser]
    @emailId NVARCHAR(255),
    @decryptedPassword NVARCHAR(50),
    @name NVARCHAR(255) OUTPUT
AS
BEGIN
    IF EXISTS (
        SELECT 1
        FROM tbl_SignUp
        WHERE emailId = @emailId
        AND decryptedPassword = @decryptedPassword
        AND name = @name
    )
    BEGIN
        SELECT
            su.id,
            su.name,
            su.emailId
        FROM tbl_SignUp su
        WHERE su.emailId = @emailId
            AND su.decryptedPassword = @decryptedPassword
            AND su.name = @name;

        -- Set the @name parameter value
        SET @name = (SELECT su.name FROM tbl_SignUp su WHERE su.emailId = @emailId AND su.decryptedPassword = @decryptedPassword AND su.name = @name);
    END
END;
GO
```
